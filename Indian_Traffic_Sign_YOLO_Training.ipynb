```python
# ╔══════════════════════════════════════════════════════════════╗
# ║  CELL 3 — Train YOLOv8n (Nano) Model                        ║
# ╚══════════════════════════════════════════════════════════════╝
# This will train the model for 25 epochs. It should take ~10-15 minutes on a T4 GPU.
from ultralytics import YOLO

# Load a pretrained YOLOv8 nano model
model = YOLO('yolov8n.pt')

yaml_path = f"{dataset.location}/data.yaml"
print(f"🚀 Training on {yaml_path}...")

# Train the model
results = model.train(
    data=yaml_path,
    epochs=25,
    imgsz=640,
    batch=16,
    project='IndianTrafficSigns',
    name='yolov8_training',
    patience=5,
)

print("\n🎉 Training Complete!")
```
